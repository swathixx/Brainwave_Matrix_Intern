```python
# 📦 Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# 📂 Step 2: Load the Dataset
df = pd.read_csv("superstore_sales_data.csv")
df.head()
```

```python
# 🔍 Step 3: Data Overview
print("Shape of dataset:", df.shape)
print("\nData Types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# 📈 Step 4: Sales by Category
category_sales = df.groupby("Category")["Total Sales"].sum().sort_values(ascending=False)
category_sales.plot(kind="bar", title="Total Sales by Category", color="skyblue")
plt.ylabel("Sales")
plt.show()
```

```python
# 🛍️ Step 5: Top 10 Selling Products
top_products = df.groupby("Product")["Total Sales"].sum().sort_values(ascending=False).head(10)
top_products.plot(kind="bar", title="Top 10 Products by Sales", color="orange")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 📍 Step 6: Sales by Location
location_sales = df.groupby("Location")["Total Sales"].sum().sort_values(ascending=False)
location_sales.plot(kind="pie", autopct='%1.1f%%', figsize=(6, 6), title="Sales Distribution by Location")
plt.ylabel("")
plt.show()
```

```python
# 📅 Step 7: Monthly Sales Trend
df['Date'] = pd.to_datetime(df['Date'])
monthly_sales = df.resample('M', on='Date')["Total Sales"].sum()
monthly_sales.plot(marker='o', title="Monthly Sales Trend", color="green")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```
